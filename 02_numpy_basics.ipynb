```python
# Import NumPy
import numpy as np
```

```python
# 1D Array (Vector)
arr = np.array([10, 20, 30, 40])
print(arr)
print(type(arr))
```

```python
# Basic Math (Where NumPy Shines)

print(arr + 5)
print(arr * 2)
```

```python
# Mean, Sum (Used Everywhere in ML)

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
```

```python
# 2D Array (Matrix – CORE FOR AI)

matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(matrix)
print("Shape:", matrix.shape)
```

```python
print(matrix[0, 0])
print(matrix[1, 2])
```

```python
# Real AI-Style Example

marks = np.array([70, 85, 90, 60, 75])

average = np.mean(marks)
passed = marks >= 70

print("Average: ", average)
print("Passed students :", passed)
```
