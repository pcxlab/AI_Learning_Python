```python

# Variables & Data Types
age = 30
height = 5.9
name = "AI Learner"
is_student = True

print(age, type(age))
print(height, type(height))
print(name, type(name))
print(is_student, type(is_student))
```

```python
# Lists (Very Important
scores = [85,90,78,92]
print(scores)
print("First score", scores[0])
print("Average score:", sum(scores) / len(scores))
```

```python
# Dictionaries (AI loves this)
student = {
    "name" : "Rahul",
    "age" : 25,
    "marks" : 88
}

print(student["name"])
print(student["marks"])

```

```python
# Loops
for score in scores:
    print(score)
```

```python
#Conditions
marks = 72

if marks >= 90:
    print("Excellent")
elif marks >= 60:
    print("Good")
else:
    print("Needs improvement")
```

```python
#Functions (CRITICAL FOR AI)
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

avg = calculate_average(scores)
print("average:", avg)
```
